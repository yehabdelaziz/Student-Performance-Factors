### Now Let's see the distribution of the student scores

```python
df['Exam_Score'].hist(bins=100, alpha=0.7, color='green', edgecolor='black')
plt.title('Distribution of exam scores')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### This looks like a gaussian/normal distribution. Let's see if we can fit it to a gaussian and find the mean and the standard deviation

```python
mu, std = stats.norm.fit(df['Exam_Score'])  # Fit a normal distribution to the data
```

```python
hist, bins = np.histogram(df['Exam_Score'], bins=100,) # Bin information

x = np.linspace(40, 100, 100)  # Generate x values
p = stats.norm.pdf(x, mu, std)  # Get the y values for the Gaussian curve
p=p*max(hist)/max(p)
plt.plot(x, p, color='red', linewidth=2,label='Gaussian Fit')  # Plot the fitted Gaussian curve
plt.title('Histogram of Age with Fitted Gaussian Curve')
plt.xlabel('Age')
plt.ylabel('Density')
plt.grid(axis='y', alpha=0.75)

df['Exam_Score'].hist(bins=100, alpha=0.7, color='green', edgecolor='black',label='Data')
plt.title('Distribution of exam scores')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.legend()
plt.text(85, 500, 'Median: {:.2f}\nMean: {:.2f}\nStd: {:.2f}'.format(np.percentile(df['Exam_Score'],50),mu, std,), fontsize=12, bbox=dict(facecolor='red', alpha=0.5))
plt.show()

```

#### The exam scores have a median of 67, a mean of 67.24 and standard deviation of 3.89
