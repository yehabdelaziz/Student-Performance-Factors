### Now Let's see the distribution of the student scores

```python
df['Exam_Score'].hist(bins=100, alpha=0.7, color='green', edgecolor='black')
plt.title('Distribution of exam scores')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### This looks like a gaussian/normal distribution. Let's see if we can fit it to a gaussian and find the mean and the standard deviation

```python
mu, std = stats.norm.fit(df['Exam_Score'])  # Fit a normal distribution to the data
```

```python
hist, bins = np.histogram(df['Exam_Score'], bins=100,) # Bin information

x = np.linspace(40, 100, 100)  # Generate x values
p = stats.norm.pdf(x, mu, std)  # Get the y values for the Gaussian curve
p=p*max(hist)/max(p)
plt.plot(x, p, color='red', linewidth=2,label='Gaussian Fit')  # Plot the fitted Gaussian curve
plt.title('Histogram of Age with Fitted Gaussian Curve')
plt.xlabel('Age')
plt.ylabel('Density')
plt.grid(axis='y', alpha=0.75)

df['Exam_Score'].hist(bins=100, alpha=0.7, color='green', edgecolor='black',label='Data')
plt.title('Distribution of exam scores')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.legend()
plt.text(85, 500, 'Median: {:.2f}\nMean: {:.2f}\nStd: {:.2f}'.format(np.percentile(df['Exam_Score'],50),mu, std,), fontsize=12, bbox=dict(facecolor='red', alpha=0.5))
plt.show()

```

### Now Let's examine some of the categorical variables in the data 

```python
# Define a function to create pie charts for each categorical variable

def plot_pie_charts(dataframe):
    categorical_columns = dataframe.select_dtypes(include='object').columns
    num_columns = len(categorical_columns)

    # Set up the grid for subplots: let's use 4 columns and calculate rows automatically
    cols = 4  # Number of columns in the grid
    rows = (num_columns + cols - 1) // cols  # Calculate the number of rows needed

    # Set up the figure
    fig, axes = plt.subplots(rows, cols, figsize=(16, 4 * rows))

    # Flatten axes to 1D if there are multiple rows
    axes = axes.flatten()

    for i, column in enumerate(categorical_columns):
        counts = dataframe[column].value_counts()
        axes[i].pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
        axes[i].set_title(f'{column}')
        axes[i].axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle

    # Hide any empty subplots (if fewer than the total grid size)
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')

    # Adjust layout
    plt.tight_layout()
    plt.show()

# Call the function to plot pie charts
plot_pie_charts(df)
```
